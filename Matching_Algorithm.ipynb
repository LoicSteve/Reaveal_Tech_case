```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

def preprocess_name(name):
    # Remove non-alphanumeric characters and convert to lowercase
    return re.sub(r'[^a-zA-Z0-9]', '', name).lower()

def match_companies(dataset_A, dataset_B, output_file):
    # Load the restructured datasets A and B into pandas DataFrames
    df_A = pd.read_csv(dataset_A)
    df_B = pd.read_csv(dataset_B)

    # Filter out companies with missing 'Company Name' values in both datasets
    df_A = df_A.dropna(subset=['Company Name'])
    df_B = df_B.dropna(subset=['Company Name'])

    # Reset the index of the DataFrames after dropping missing values
    df_A = df_A.reset_index(drop=True)
    df_B = df_B.reset_index(drop=True)

    # Preprocess company names
    df_A['Preprocessed Name'] = df_A['Company Name'].apply(preprocess_name)
    df_B['Preprocessed Name'] = df_B['Company Name'].apply(preprocess_name)

    # Create TF-IDF vectors for company names in both datasets
    tfidf_vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 3))
    tfidf_matrix_A = tfidf_vectorizer.fit_transform(df_A['Preprocessed Name'])
    tfidf_matrix_B = tfidf_vectorizer.transform(df_B['Preprocessed Name'])

    # Calculate cosine similarity between TF-IDF vectors of dataset A and B
    cosine_similarities = linear_kernel(tfidf_matrix_A, tfidf_matrix_B)

    # Create a new DataFrame to store the matched pairs and relevant information
    matched_df = pd.DataFrame(columns=['Company_A_ID', 'Company_A_Name', 'Company_B_ID', 'Company_B_Name', 'Similarity_Score'])

    # Iterate through each company in dataset A and find its best match in dataset B
    for index_A, company_A in df_A.iterrows():
        best_match_index = cosine_similarities[index_A].argmax()
        best_match_score = cosine_similarities[index_A, best_match_index]

        # Set a threshold for similarity score
        threshold = 0.8

        # If similarity score is above the threshold, consider them a match
        if best_match_score >= threshold:
            best_match = df_B.iloc[best_match_index]
            matched_pair_info = {
                'Company_A_ID': company_A['Company ID'],
                'Company_A_Name': company_A['Company Name'],
                'Company_B_ID': best_match['Company ID'],
                'Company_B_Name': best_match['Company Name'],
                'Similarity_Score': best_match_score,
                # Add any other relevant information about the match
            }
            matched_df = matched_df.append(matched_pair_info, ignore_index=True)

    # Save the matched pairs to a new CSV file
    matched_df.to_csv(output_file, index=False)

if __name__ == "__main__":
    dataset_A_file = "./restructured_datasetA.csv"
    dataset_B_file = "./restructured_datasetB.csv"
    output_match_file = "./matched_companiesF.csv"

    match_companies(dataset_A_file, dataset_B_file, output_match_file)
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

def preprocess_name(name):
    # Remove non-alphanumeric characters and convert to lowercase
    return re.sub(r'[^a-zA-Z0-9]', '', name).lower()

def preprocess_postal_code(postal_code):
    # Convert postal code to string and remove non-alphanumeric characters
    return re.sub(r'[^a-zA-Z0-9]', '', str(postal_code)).lower()

def preprocess_country(country):
    # Convert country name to lowercase
    return str(country).lower()

def match_companies(dataset_A, dataset_B, output_file):
    # Load the restructured datasets A and B into pandas DataFrames
    df_A = pd.read_csv(dataset_A)
    df_B = pd.read_csv(dataset_B)

    # Filter out companies with missing 'Company Name' values in both datasets
    df_A = df_A.dropna(subset=['Company Name'])
    df_B = df_B.dropna(subset=['Company Name'])

    # Reset the index of the DataFrames after dropping missing values
    df_A = df_A.reset_index(drop=True)
    df_B = df_B.reset_index(drop=True)

    # Preprocess company names, postal codes, and countries
    df_A['Preprocessed Name'] = df_A['Company Name'].apply(preprocess_name)
    df_B['Preprocessed Name'] = df_B['Company Name'].apply(preprocess_name)

    df_A['Preprocessed Postal Code'] = df_A['Postal Code'].apply(preprocess_postal_code)
    df_B['Preprocessed Postal Code'] = df_B['Postal Code'].apply(preprocess_postal_code)

    df_A['Preprocessed Country'] = df_A['Country'].apply(preprocess_country)
    df_B['Preprocessed Country'] = df_B['Country'].apply(preprocess_country)

    # Create TF-IDF vectors for company names, postal codes, and countries in both datasets
    tfidf_vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 3))
    tfidf_matrix_A = tfidf_vectorizer.fit_transform(df_A['Preprocessed Name'] + df_A['Preprocessed Postal Code'] + df_A['Preprocessed Country'])
    tfidf_matrix_B = tfidf_vectorizer.transform(df_B['Preprocessed Name'] + df_B['Preprocessed Postal Code'] + df_B['Preprocessed Country'])

    # Calculate cosine similarity between TF-IDF vectors of dataset A and B
    cosine_similarities = linear_kernel(tfidf_matrix_A, tfidf_matrix_B)

    # Create a new DataFrame to store the matched pairs and relevant information
    matched_df = pd.DataFrame(columns=['Company_A_ID', 'Company_A_Name', 'Company_B_ID', 'Company_B_Name', 'Similarity_Score'])

    # Iterate through each company in dataset A and find its best match in dataset B
    for index_A, company_A in df_A.iterrows():
        best_match_index = cosine_similarities[index_A].argmax()
        best_match_score = cosine_similarities[index_A, best_match_index]

        # Set a threshold for similarity score
        threshold = 0.8

        # If similarity score is above the threshold, consider them a match
        if best_match_score >= threshold:
            best_match = df_B.iloc[best_match_index]
            matched_pair_info = {
                'Company_A_ID': company_A['Company ID'],
                'Company_A_Name': company_A['Company Name'],
                'Company_B_ID': best_match['Company ID'],
                'Company_B_Name': best_match['Company Name'],
                'Similarity_Score': best_match_score,
                # Add any other relevant information about the match
            }
            matched_df = matched_df.append(matched_pair_info, ignore_index=True)

    # Save the matched pairs to a new CSV file
    matched_df.to_csv(output_file, index=False)

if __name__ == "__main__":
    dataset_A_file = "./restructured_datasetA.csv"
    dataset_B_file = "./restructured_datasetB.csv"
    output_match_file = "./matched_companiesFF.csv"

    match_companies(dataset_A_file, dataset_B_file, output_match_file)
```
